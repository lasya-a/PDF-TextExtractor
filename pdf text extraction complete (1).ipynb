```python
def extract_director_info(pdf_file):
    # Open the PDF file
    pdf_document = fitz.open(pdf_file)
    
    director_info = {
        "Name": None,
        "Director_Type": None,
        "DIN": None
    }
    
    # Iterate through each page of the PDF
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text = page.get_text()
        
        # Search for director information patterns
        name_match = re.search(r'Name: (.+)', text)
        type_match = re.search(r'Director Type: (.+)', text)
        din_match = re.search(r'DIN: (\d+)', text)
        
        if name_match:
            director_info["Name"] = name_match.group(1)
        if type_match:
            director_info["Director_Type"] = type_match.group(1)
        if din_match:
            director_info["DIN"] = din_match.group(1)
    
    return director_info
```
